```python
import pandas as pd
import sqlite3
```

# Connect sql database

```python
name_bd = '../checking-logs.sqlite.sqlite_copy'
sqlite_connection = sqlite3.connect(name_bd)
pd.io.sql.read_sql('PRAGMA table_info(pageviews);', sqlite_connection)
```

# Get first 10 lines

```python
pd.io.sql.read_sql("select * from pageviews LIMIT 10", sqlite_connection)
```

```python
pd.io.sql.read_sql("SELECT uid, datetime FROM pageviews WHERE uid LIKE 'user_%' AND uid NOT LIKE '%admin%' ORDER BY uid ASC", sqlite_connection)
```

```python
pageviews = pd.io.sql.read_sql("SELECT uid, datetime \
                   FROM pageviews WHERE uid LIKE 'user_%' \
                   ORDER BY uid ASC", sqlite_connection,
                  index_col='datetime', parse_dates=['datetime'])
pageviews
```
